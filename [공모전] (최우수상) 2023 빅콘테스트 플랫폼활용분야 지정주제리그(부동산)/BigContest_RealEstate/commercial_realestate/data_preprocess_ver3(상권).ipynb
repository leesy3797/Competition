```python
import pandas as pd
```

# 상권코드 merge

```python
# 상권데이터
path0 = '/content/drive/MyDrive/머신러닝/ML_팀프로젝트/0911 원본 조합하기/상권합본.csv'
sg_code = pd.read_csv(path0)
```

```python
# 결합용 대회데이터
path1 = '/content/drive/MyDrive/머신러닝/ML_팀프로젝트/0911 원본 조합하기/상권코드결합용_230911_한석.csv'
main_code = pd.read_csv(path1)
```

```python
# 상권데이터 확인
sg_code.head(1)
```

```python
# 결합용 대회데이터  => 확인

main_code.head(10)
```

```python
# 결합용 대회데이터 => 월을 기준으로 '분기' 피처 추가

def to_quarter( x ):
    if x in ['01', '02', '03']:
        return 1
    elif x in ['04', '05', '06']:
        return 2
    elif x in ['07', '08', '09']:
        return 3
    else:
        return 4
    pass

main_code['기준_분기_코드'] = main_code['DATA_CRTR_YM'].apply(lambda x : to_quarter(str(x)[-2:]))
```

```python
# 상권데이터 => [상권_코드] 피처 앞에 '2'를 추가(년도마다 맨 앞자리 숫자가 다르기 때문 - 현 상권데이터 상에는 맨 앞자리 숫자가 빠진 상태)

sg_code['상권_코드'] = '2' + sg_code['상권_코드'].astype(str)
sg_code['상권_코드'].astype(int)
```

```python
# 결합용 대회 데이터 => 결합 이전 피처명 통일

main_code = main_code.rename(columns={'TRDAR_CD': '상권_코드'})
```

```python
main_code.info()
```

```python
sg_code.info()
```

```python
sg_code.drop(['Unnamed: 0'], axis=1, inplace=True)
sg_code.info()
```

```python
# 결합 이전 데이터 타입 통일

main_code['상권_코드'] = main_code['상권_코드'].astype(int)
main_code['기준_분기_코드'] = main_code['기준_분기_코드'].astype(int)
sg_code['상권_코드'] = sg_code['상권_코드'].astype(int)
sg_code['기준_분기_코드'] = sg_code['기준_분기_코드'].astype(int)
```

```python
# 결합용 대회 데이터, 상권데이터 => 확인

display(main_code.head(2)), display(sg_code.head(2))
```

```python
# 결합용 대회 데이터, 상권데이터 => '상권_코드','기준_분기_코드'를 기준으로 merge
# 1차 결합된 데이터 => merged_df

merged_df = pd.merge(main_code, sg_code, on=['상권_코드', '기준_분기_코드'], how='left')
merged_df
```

# 법정동 merge

```python
# 1차 결합 데이터 => 결합 이전 피처명 통일

merged_df = merged_df.rename(columns={'STDG_EMD_CD': '법정동_코드'})
```

```python
merged_df.info()
```

```python
# 상권데이터 => bub 변수에 저장

bub = sg_code.copy()
```

```python
# 상권데이터 => 확인
bub.info()
```

```python
merged_df.head(1)
```

```python
# 중복된 법정동_코드 열 삭제(마지막 열)

merged_df = merged_df.iloc[:, :-1]
```

```python
# 삭제 되었는지 확인

merged_df.info()
```

```python
merged_df['법정동_코드']
```

```python
bub['법정동_코드']
```

```python
# 상권데이터 => 법정동_코드, 기준_분기_코드를 기준으로 평균 집계

bub_new = bub.groupby(['법정동_코드', '기준_분기_코드']).mean().reset_index()
```

```python
# 1차 결합 데이터, 상권데이터 => 법정동_코드, 기준_분기_코드 기준으로 merge
# 2차 결합 데이터 => merged_df2

merged_df2 = pd.merge(merged_df, bub_new, on=['법정동_코드', '기준_분기_코드'], how='left')
merged_df2
```

```python
# forfinal_df 변수에 2차 결합 데이터 저장

forfinal_df = merged_df2
```

```python
forfinal_df.shape
```

```python
forfinal_df.info()
```

```python
# 2차 결합 데이터 => 중복된 피처(x,y) => x의 null값에 y의 값을 채워줌

for column in forfinal_df.columns:
    if column.endswith('_x'):
        corresponding_column = column.replace('_x', '_y')
        forfinal_df[column].fillna(forfinal_df[corresponding_column], inplace=True)

forfinal_df[['총 상주인구 수_x', '총 상주인구 수_y']]
```

```python
forfinal_df.info()
```

```python
# 2차 결합 데이터 => '_y'로 끝나는 피처 제거

for column in forfinal_df.columns:
    if column.endswith('_y'):
        forfinal_df.drop(column, axis=1, inplace=True)
forfinal_df.info()
```

```python
# 2차 결합 데이터 => 피처명에서 '_x' 제거

for column in forfinal_df.columns:
    if column.endswith('_x'):
        new_column_name = column[:-2]  # '_x'를 제거
        forfinal_df.rename(columns={column: new_column_name}, inplace=True)
forfinal_df.info()
```

```python
forfinal_df = forfinal_df.rename(columns={'월_평균_소득_금액': '상권_월_평균_소득_금액'})
forfinal_df = forfinal_df.rename(columns={'지출_총금액': '상권_지출_총금액'})
forfinal_df = forfinal_df.rename(columns={'식료품_지출_총금액': '상권_식료품_지출_총금액'})

forfinal_df = forfinal_df.rename(columns={'운영_영업_개월_평균': '상권_운영_영업_개월_평균'})
forfinal_df = forfinal_df.rename(columns={'폐업_영업_개월_평균': '상권_폐업_영업_개월_평균'})

forfinal_df = forfinal_df.rename(columns={'총 상주인구 수': '상권_총 상주인구 수'})
forfinal_df = forfinal_df.rename(columns={'남성 상주인구 수': '상권_남성 상주인구 수 '})
forfinal_df = forfinal_df.rename(columns={'여성 상주인구 수': '상권_여성 상주인구 수 '})

forfinal_df = forfinal_df.rename(columns={'총 가구 수': '상권_총 가구 수'})
forfinal_df = forfinal_df.rename(columns={'아파트 가구 수': '상권_아파트 가구 수'})
forfinal_df = forfinal_df.rename(columns={'비 아파트 가구 수': '상권_비 아파트 가구 수'})

forfinal_df = forfinal_df.rename(columns={'총_생활인구_수': '상권_총_생활인구_수'})
forfinal_df = forfinal_df.rename(columns={'남성_생활인구_수': '상권_남성_생활인구_수'})
forfinal_df = forfinal_df.rename(columns={'여성_생활인구_수': '상권_여성_생활인구_수'})
forfinal_df = forfinal_df.rename(columns={'집객시설_수': '상권_집객시설_수'})
```
