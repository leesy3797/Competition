```python
pilji_mapping = pilji_mapping.rename(columns = {'TRDAR_CD':'상권_코드', 'STDG_EMD_CD' : '법정동_코드'})
pilji_mapping.head()
```

```python
def to_quarter( x ):
    if x in ['01', '02', '03']:
        return 1
    elif x in ['04', '05', '06']:
        return 2
    elif x in ['07', '08', '09']:
        return 3
    else:
        return 4
    pass

pilji_mapping['기준_분기_코드'] = pilji_mapping['DATA_CRTR_YM'].apply(lambda x : to_quarter(str(x)[-2:]))
```

```python
pilji_mapping.head(), pilji_mapping.shape, pilji_mapping.info()
```

```python
merge_first = pilji_mapping.merge(merger7, how = 'left', on = ['상권_코드', '기준_분기_코드'], suffixes= ('', '_right') )
```
